```python
import matplotlib.pyplot as plt
x=[25,26,27,28,29]
y=[2000,3000,4000,5000,6000]
y1=[3000,4000,5000,6000,7000]
plt.plot(x,y,marker='o',linestyle='--',color='y',label="y_axis")
plt.plot(x,y1,marker='s',linestyle='-.',color='k',label="y1_axis")
plt.xlabel('ages')
plt.ylabel('salary')
plt.title('Linear plot')
plt.tight_layout()
plt.legend()
plt.style.use('fivethirtyeight')
#print(plt.style.available)
plt.show()
```

```python
import pandas as pd
m=pd.read_csv('mpg.csv')
m.head()
```

```python
#bar graph
import matplotlib.pyplot as plt
plt.bar(m['year'],m['cty'])
plt.show()
```

```python
#hist graph
import matplotlib.pyplot as plt
plt.hist(m['year'])
plt.show()
```

```python
#scatter graph
import matplotlib.pyplot as plt
plt.scatter(m['hwy'],m['cty'])
plt.show()
```

```python
#pie graph
import matplotlib.pyplot as plt
labels=['python','html','sql']
slices=[65345,75654,77888]
explode=[0,0,0.1]
plt.pie(slices,labels=labels,explode=explode)
plt.show()
```
